```python
import ee
import geemap
import numpy as np
import geopandas as gpd
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Conv2D, LSTM, Dense, Flatten, Dropout, TimeDistributed, Reshape, Conv2DTranspose,BatchNormalization, Activation, ConvLSTM2D
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import rasterio
from tensorflow.keras.callbacks import Callback
from tqdm import tqdm
import time
from tensorflow.keras import backend as K
import os
from tensorflow.config.experimental import set_virtual_device_configuration, VirtualDeviceConfiguration
import tensorflow.keras as keras
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix
import keras_tuner as kt
```

```python
# Authenticate and initialize Earth Engine
ee.Authenticate()
ee.Initialize()
```

```python
# Define a smaller forest region in central Ireland (Kinnitty forest)
kinnitty_coords = [
    [-7.745565, 52.988778],
    [-7.745565, 53.157092],
    [-7.437885, 53.157092],
    [-7.437885, 52.988778],
    [-7.745565, 52.988778]
]

# Create a polygon from the coordinates for use with GEE
kinnitty = ee.Geometry.Polygon([kinnitty_coords])
```

```python
# Create map
Map = geemap.Map()

# Center the map on Kinnitty
Map.centerObject(kinnitty, zoom=11)

# Add Kinnitty polygon to the map
Map.addLayer(kinnitty, {'color': 'red'}, 'Kinnitty')

Map
```

```python
# Load Global Forest Change dataset
gfc_dataset = ee.Image('UMD/hansen/global_forest_change_2022_v1_10')

# Extract tree loss data
tree_loss = gfc_dataset.select('lossyear').clip(kinnitty)

#create a list of years to loop through
years = list(range(2001, 2023))

#Create a dictionary to hold each years loss raster
tree_loss_by_year = {}

#loop through each year and add each raster to the dictionary
for year in years:
    yearly_loss = tree_loss.eq(year - 2000)
    tree_loss_by_year[year] = yearly_loss
```

```python
#Create empty arrays to store data for each raster for use in model
arrays = []
transforms = []
crs_list = []

for year in years:
   
    yearly_loss = tree_loss_by_year[year].rename('loss')

    # get the loss band for each year and converty to numpy array
    arr = geemap.ee_to_numpy(yearly_loss, bands=['loss'], region=kinnitty, scale=30)

    # get the crs and transform data for each image which is required later for mapping with GEE
    info = yearly_loss.getInfo()
    crs = info['bands'][0]['crs']
    transform = info['bands'][0]['crs_transform']

    arrays.append(arr)
    transforms.append(transform)
    crs_list.append(crs)

# Stack arrays
data = np.stack(arrays, axis=0)  # shape: (num_years, height, width, 1)

#print shape to see that it is in correct format
print("Data shape:", data.shape)
```

```python
# Prepare sequences for model input
sequence_length = 3
num_samples = data.shape[0] - sequence_length

X = []
y = []

for i in range(num_samples):
    X.append(data[i:i+sequence_length])
    y.append(data[i+sequence_length])

X = np.array(X)  # shape: (num_samples, sequence_length, height, width, 1)
y = np.array(y)  # shape: (num_samples, height, width, 1)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Split data into training and testing sets
train_size = int(0.8 * num_samples)

X_train = X[:train_size]
y_train = y[:train_size]

X_test = X[train_size:]
y_test = y[train_size:]

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
#Create function to build model with input parameters from tuner
def build_model(hp):
    K.clear_session()
    model = Sequential()
    
    num_conv_lstm_layers = 1  
    
    for i in range(num_conv_lstm_layers):
        filters = hp.Int(f'filters_{i}', min_value=16, max_value=32, step=16)
        kernel_size = hp.Choice(f'kernel_size_{i}', values=[3])
        model.add(ConvLSTM2D(
            filters=filters,
            kernel_size=(kernel_size, kernel_size),
            input_shape=(sequence_length, data.shape[1], data.shape[2], 1),
            padding='same',
            return_sequences=False
        ))
        
        if hp.Boolean(f'use_batchnorm_{i}'):
            model.add(BatchNormalization())
    
    # Final Conv2D layer
    model.add(Conv2D(
        filters=1,
        kernel_size=(1, 1),
        activation='sigmoid',
        padding='same'
    ))
    
    # Compile model
    optimizer = hp.Choice('optimizer', ['adam'])
    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

```python
#Define the Hyperparameter tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=7,  # Fewer epochs
    factor=2,
    directory='hyperparam_tuning',
    project_name='tree_loss_prediction'
)
```

```python
# Early stopping callback
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=1)

# Run the hyperparameter search
tuner.search(
    X_train,
    y_train,
    epochs=6,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping]
)
```

```python
# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the best model
model = tuner.hypermodel.build(best_hps)
```

```python
# Display the best hyperparameters
for param in best_hps.values.keys():
    print(f"{param}: {best_hps.get(param)}")
```

```python
# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping]
)
```

```python
# Threshold predictions to get binary outputs (tree loss = 1)
y_pred_binary = (y_pred > 0.4844389).astype(np.float32)
```

```python
#Get crs and transform for Kinnitty forest to act as boundary to display predictions on GEEmap

# Coordinates for Kinnitty
kinnitty_coords = [
    [-7.745565, 52.988778],  # Lower-left
    [-7.745565, 53.157092],  # Upper-left
    [-7.437885, 53.157092],  # Upper-right
    [-7.437885, 52.988778]   # Lower-right
]

# get bounds from coords
xmin = min(coord[0] for coord in kinnitty_coords)
xmax = max(coord[0] for coord in kinnitty_coords)
ymin = min(coord[1] for coord in kinnitty_coords)
ymax = max(coord[1] for coord in kinnitty_coords)

#have to hardcode adjust bounds to resolve issues with incorrect bounds
geo_width = xmax - xmin
xmax = xmin + 1.83 * geo_width
geo_height = ymax - ymin
ymin = ymax - 0.55 * geo_height  

num_pixels_x = data.shape[2]  # Number of pixels in original width
num_pixels_y = data.shape[1]  # Number of pixels in original height

# Recalculate pixel sizes based on new bounds
pixel_size_x = (xmax - xmin) / num_pixels_x
pixel_size_y = (ymax - ymin) / num_pixels_y

# Update the transform array with new bounds
transform = [pixel_size_x, 0, xmin, 0, -pixel_size_y, ymax]

# Set CRS
crs = 'EPSG:4326'

# Print out the new transform and CRS for verification
print("Adjusted CRS:", crs)
print("Adjusted Transform:", transform)
```

```python
# Create new map
Map = geemap.Map()

# Center map on the kinnitty
Map.centerObject(kinnitty, zoom=11)

#loop through each prediction
for i in range(len(y_pred)):
    #reshape data to numpy array
    pred_array = y_pred_binary[i].reshape(data.shape[1], data.shape[2])
    actual_array = y_test[i].reshape(data.shape[1], data.shape[2])

    #convert numpy arrays to Earth Engine images
    pred_image = geemap.numpy_to_ee(pred_array, crs=crs, transform=transform)
    actual_image = geemap.numpy_to_ee(actual_array, crs=crs, transform=transform)

    #apply mask to the images to only show tree loss for each layer
    pred_image_masked = pred_image.updateMask(pred_image.eq(1))
    actual_image_masked = actual_image.updateMask(actual_image)
    
    #add layers to the map
    Map.addLayer(pred_image_masked, {'min': 0, 'max': 1, 'palette': ['red']}, f'Predicted Tree Loss {i}')
    Map.addLayer(actual_image_masked, {'min': 0, 'max': 1, 'palette': ['00FF00']}, f'Actual Tree Loss {i}')

Map
```
